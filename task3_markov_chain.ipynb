```python
import markovify

text = """
Artificial intelligence is transforming industries across the world.
Machine learning allows computers to learn patterns from data.
Deep learning models can analyze complex information.
Natural language processing helps machines understand human language.
AI is widely used in healthcare, finance, and automation.
Generative AI systems can create text, images, and music.
Data science combines statistics, programming, and machine learning.
Artificial intelligence will continue to shape the future of technology.
AI models are improving search engines and recommendation systems.
Robotics and AI are working together in modern manufacturing.
"""

model = markovify.Text(text, state_size=1)

for i in range(5):
    print(model.make_sentence())
```
